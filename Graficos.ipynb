Prefira usar graficos de barra, de preferencia sem cores.

```python
# Create horizontal bars
plt.figure(figsize=(8, 6))

total_count = species_counts.sum()
percentages = (species_counts / total_count) * 100

# barras verticais
bars = plt.bar(x=species_counts.index, height=species_counts.values)
plt.xlabel("Especies")
plt.ylabel("Frequencia")


# Add percentage values to bars
for bar, percentage in zip(bars, percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{percentage:.1f}%', va='bottom', ha='center')
    #f'{percentage:.1f}%' formata a porcentagem com uma casa decimal

plt.show()
```

Quando vamos comparar 2 variveis categorias, geralmente recorremos as tabelas de contingencia.

```python
# Criar uma tabela de contingncia (cross-tab) para categorias
cross_tab = pd.crosstab(df.species, df.sex,  normalize='index') # comparando os sexos pelas especies,(normalize=all: todos, index:linha, columns: coluna)
cross_tab
```

Podemos usar as barras para representao grfica. para at duas classes, podemos usar barras empilhadas

```python
from matplotlib.ticker import PercentFormatter



# Plotar as barras empilhadas
ax=cross_tab.plot(kind='bar', stacked=True,width=0.8,figsize=(8,6))


# Adicionar rtulos e ttulo
plt.ylabel('Proporo')
plt.xlabel('Categoria')
ax.yaxis.set_major_formatter(PercentFormatter(xmax=1)) #porcentagem
plt.xticks(rotation=40)
for p in ax.patches:
    width = p.get_width()    # Largura da barra
    height = p.get_height()  # Altura da barra (proporo)
    x, y = p.get_xy()        # Posio da barra
    if height > 0:           # Adicionar porcentagem se houver altura
        ax.text(x + width / 2, y + height / 2, f'{height:.1%}',
                ha='center', va='center', color='white', fontsize=12)

# Exibir o grfico

plt.show()
```

Mais de duas classes recomenda se usar barras paralelas

```python
# Criar uma tabela de contingncia (cross-tab) para categorias
cross_tab = pd.crosstab(df.sex, df.species,  normalize='index')
cross_tab = cross_tab[cross_tab.sum(axis=0).sort_values(ascending=False).index] # cuidar a ordem


ax=cross_tab.plot(kind='bar', stacked=False,width=0.5, figsize=(8,6) )
ax.legend(title='Especies', bbox_to_anchor=(1.05, 1), loc='upper left') # posio da legenda
ax.yaxis.set_major_formatter(PercentFormatter(xmax=1))
plt.ylabel('Proporo')
plt.xlabel('Sexo')
plt.xticks(rotation=0)

for p in ax.patches:
    width = p.get_width()    # Largura da barra
    height = p.get_height()  # Altura da barra (proporo)
    x, y = p.get_xy()        # Posio da barra
    if height > 0:           # Adicionar porcentagem se houver altura
        ax.text(x + width / 2, y + height / 2, f'{height:.1%}',
                ha='center', va='center', color='white', fontsize=10)
plt.tight_layout()
plt.show()
```

Outra forma recorrente de grficos, ao series temporais, note que vamos usar grade de fundo.

```python
#from re import template
import numpy as np

pontos = 100 # numero de pontos

# geranto tempo (sequencia)
time = np.arange(pontos)

# vamos gerar pontos aleatorios

temp = np.random.uniform(20, 35, pontos)

P = np.random.uniform(10, 15, pontos)

RH= np.random.uniform(50, 100, pontos)

data = {
    'Time': time,
    'Temperatura': temp,
    'Presso': P,
    'Umidade': RH
}
df3 = pd.DataFrame(data)
```

```python

# Configurar o tema do grfico
plt.style.use('seaborn-v0_8-white')  # Usar um tema branco com grades
plt.figure(figsize=(16, 6))  # Definir o tamanho da figura

# Criar e exibir o grfico de linhas
ax = df3.plot(x='Time',
             kind='line',
             grid=True,  # Adicionar uma grade no fundo
             linewidth=2,  #  largura da linha
             alpha=0.8,  # opacidade da linha
             ax=plt.gca() )


plt.xlabel('Tempo', fontsize=18)  # Tamanho do rtulo do eixo x
plt.ylabel('valor', fontsize=18)  # Tamanho do rtulo do eixo y
plt.tick_params(axis='both', labelsize=16)

# Personalizar a legenda
plt.legend(loc='upper right',
           bbox_to_anchor=(1.15, 1),
           fontsize=12  # Tamanho da fonte da legenda
           )

# Exibir o grfico
plt.show()
```

```python
import string
num_vars = df3.shape[1] - 1  # Subtrai 1 porque 'Time' no  uma varivel de interesse

# Criar um painel de subgrficos
fig, axes = plt.subplots(nrows=num_vars, ncols=1, figsize=(10, 6), sharex=True)
# definir unidades
units = {'Temperatura': 'C', 'Presso': 'Pa', 'Umidade': '%'}

# Iterar sobre as colunas do DataFrame e criar um grfico para cada uma
for i, col in enumerate(df3.columns[1:]):  # Ignora a primeira coluna ('Time')
    ax = axes[i]
    ax.plot(df3['Time'], df3[col], linewidth=2, marker='o')  # Ajuste o estilo conforme necessrio
    ax.set_ylabel(col, fontsize=14)  # Ttulo para cada grfico
    letter = string.ascii_lowercase[i + 1]  # 'a'  o ndice 0, 'b'  1, etc.
    ax.grid(True)  # Adiciona uma grade ao fundo
    plt.text(0.01, 0.95, f'({letter.casefold()})', transform=ax.transAxes, fontsize=12, fontweight='bold', va='top')
    #unidades
    unit = units.get(col, '')
    ax.set_ylabel(f"{col} ({unit})", fontsize=14)


# Configuraes gerais
plt.xlabel('Time (s)', fontsize=14)

plt.show()
```
